# Pandas Series

The first main data type we will learn about for pandas is the __Series data type__. 

A Series is very similar to a NumPy array (in fact it is built on top of the NumPy array object). What differentiates the NumPy array from a Series:
- Series can have __axis labels__, meaning it can be indexed by a label, instead of just a number location. 
- It also doesn't need to hold numeric data, it __can hold any arbitrary Python Object__.

We can define it as a <u>One-dimensional ndarray (NumPy Array) with axis lables</u>

## Reading images

```python
# === Cell 1: Read and display an image (test.jpg) ===
from PIL import Image
import matplotlib.pyplot as plt

# Read the image from the same folder
img_rgb = Image.open('test.jpg').convert('RGB')

plt.figure()
plt.imshow(img_rgb)
plt.axis('off')
plt.title('Loaded Image: test.jpg')
plt.show()
```

```python
# === Cell 2: Convert image to grayscale and show as a matrix ===
import numpy as np

img_gray = img_rgb.convert('L')  # 'L' mode = grayscale
A = np.array(img_gray)

print('Image shape (H, W):', A.shape)
print('Data type:', A.dtype)
print('Top-left 8×8 pixel values:\n', A[:8, :8])
print('Min intensity:', A.min(), ' | Max intensity:', A.max())

plt.figure()
plt.imshow(A, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title('Grayscale Image as Matrix')
plt.show()
```

```python
# === Cell 3: Normalize pixel values (0–1 range) ===
A_f = A.astype('float32') / 255.0
print('Float dtype:', A_f.dtype)
print('Range:', A_f.min(), '-', A_f.max())
print('Top-left 5×5 normalized:\n', np.round(A_f[:5, :5], 3))
```

```python
# === Cell 4: Histogram of grayscale intensities ===
plt.figure()
plt.hist(A.flatten(), bins=32, color='gray', edgecolor='black')
plt.title('Histogram of Pixel Intensities')
plt.xlabel('Intensity (0–255)')
plt.ylabel('Frequency')
plt.show()
```

```python
# === Cell 5: Simple edge detection (Sobel filters using NumPy) ===
import numpy as np
import matplotlib.pyplot as plt

P = np.pad(A.astype('float32'), 1, mode='edge')

Kx = np.array([[-1, 0, 1],
               [-2, 0, 2],
               [-1, 0, 1]], dtype='float32')
Ky = np.array([[-1, -2, -1],
               [ 0,  0,  0],
               [ 1,  2,  1]], dtype='float32')

Gx = (Kx[0,0]*P[:-2, :-2] + Kx[0,1]*P[:-2,1:-1] + Kx[0,2]*P[:-2,2:] +
      Kx[1,0]*P[1:-1, :-2] + Kx[1,1]*P[1:-1,1:-1] + Kx[1,2]*P[1:-1,2:] +
      Kx[2,0]*P[2:, :-2] + Kx[2,1]*P[2:,1:-1] + Kx[2,2]*P[2:,2:])

Gy = (Ky[0,0]*P[:-2, :-2] + Ky[0,1]*P[:-2,1:-1] + Ky[0,2]*P[:-2,2:] +
      Ky[1,0]*P[1:-1, :-2] + Ky[1,1]*P[1:-1,1:-1] + Ky[1,2]*P[1:-1,2:] +
      Ky[2,0]*P[2:, :-2] + Ky[2,1]*P[2:,1:-1] + Ky[2,2]*P[2:,2:])

edges = np.hypot(Gx, Gy)
edges = (255 * edges / edges.max()).astype('uint8')

plt.figure()
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.title('Edge Detection (Sobel)')
plt.show()
```

```python
# === Cell 6: Otsu thresholding (automatic global threshold) ===
import numpy as np
import matplotlib.pyplot as plt

# A is the grayscale matrix from earlier cells
hist = np.bincount(A.ravel(), minlength=256).astype(np.float64)
p = hist / hist.sum()

omega = np.cumsum(p)                        # cumulative class probabilities
mu = np.cumsum(p * np.arange(256))          # cumulative class means
mu_t = mu[-1]

# Between-class variance
sigma_b2 = (mu_t * omega - mu)**2 / (omega * (1.0 - omega) + 1e-12)
t = int(np.nanargmax(sigma_b2))             # Otsu threshold

mask = (A >= t)                             # foreground = True
fg_ratio = mask.mean()

print('Otsu threshold:', t)
print('Foreground ratio:', round(float(fg_ratio)*100, 2), '%')

plt.figure()
plt.imshow(mask.astype(np.uint8) * 255, vmin=0, vmax=255)
plt.axis('off')
plt.title('Binary mask (Otsu)')
```
